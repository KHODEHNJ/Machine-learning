```
import kagglehub

# Download latest version
path = kagglehub.dataset_download("solomonameh/spotify-music-dataset")

print("Path to dataset files:", path)
```

```
import os
file = os.listdir(path)
print(file)
```

```
os.path.join(path, file[0])
```

```
import pandas as pd
df = pd.read_csv(os.path.join(path, file[0]))
df.head()
```

```
df.info()
```

```
import matplotlib.pyplot as plt
x = df["playlist_genre"]
```

```
import matplotlib.pyplot as plt


genre_counts = x.head(500).value_counts()

plt.figure(figsize=(10,15))
plt.pie(genre_counts, labels=genre_counts.index, autopct='%1.1f%%', startangle=140)
plt.title("Distribution of Playlist Genre (Top 500)")
plt.axis('equal')
plt.show()
```

```

y = df['energy']
e_counts = y.head(20).value_counts()

plt.figure(figsize=(8, 8))
plt.pie(
    e_counts,
    labels=e_counts.index,
    autopct='%1.1f%%',
    startangle=140
)
plt.title("Distribution of Energy Values (Top 100)")
plt.axis('equal')
plt.show()
```

```
z =df['track_artist'].value_counts()
z = z.sort_values(ascending=False)
```

```
# فقط ۱۰ خواننده اول
z.head(100).plot(kind='barh', figsize=(10, 25), color='skyblue')
plt.title("Top 100 Artists by Number of Songs")
plt.xlabel("Artist")
plt.ylabel("Number of Songs")
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
